# NeuroData Publication Update Protocol
This notebook utilizes `pandarize` to parse pubs.bib file from neurodata.io and update the list

```python
from pandarize.frame import Pandarizer
import pandarize
import pandas as pd
pandarize.__version__
```

```python
pdr = Pandarizer()
pdr.load('https://raw.githubusercontent.com/neurodata/neurodata.io/deploy/content/bibs/pubs.bib')
# pdr.load('pubs.bib') #for local file
pdr.fit()
```

```python
filepath = 'dev/intermediate.xlsx'
ufilepath = 'dev/intermediate-updated.xlsx'
```

```python
pdr.df.to_excel(filepath, index=False, engine='xlsxwriter')
```

#### Check for duplicates

```python
pdr.df[pdr.df['alias'].duplicated()]
```

#### Make changes to the downloaded and converted file

```python
pdr.df = pd.read_excel(ufilepath, dtype=object, na_filter=False)
```

```python
# pdr.df.priority = pdr.df.priority.astype(float)
pdr.df.sort_values(['keywords'], ascending=True, inplace=True)
```
